```python
from google.colab import files
uploaded = files.upload()
```

```python
import hashlib

def calculate_hash(filename):
    sha256 = hashlib.sha256()
    with open(filename, "rb") as f:
        while chunk := f.read(4096):
            sha256.update(chunk)
    return sha256.hexdigest()

filename = list(uploaded.keys())[0]
original_hash = calculate_hash(filename)
print("Original Hash:", original_hash)
```

```python
# Upload the same file again after modifying
uploaded_new = files.upload()
filename_new = list(uploaded_new.keys())[0]
new_hash = calculate_hash(filename_new)

print("New Hash:", new_hash)
print("File changed?" , new_hash != original_hash)
```
